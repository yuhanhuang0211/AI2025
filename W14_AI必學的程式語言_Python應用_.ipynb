### Lab‚ÄØ3¬†Bar Chart‚ÄØ101¬†‚Äì¬†Titanic Passenger Count

**Objectives**

- Group data with `groupby`
- Draw a bar chart and add value labels

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# 1¬†Group by class
class_cnt = titanic.groupby("class")["survived"].count()

# 2¬†Plot
plt.figure(figsize=(6,4))
bars = plt.bar(class_cnt.index, class_cnt.values, color=["#1f77b4", "#ff7f0e", "#2ca02c"])
plt.title("Passenger Count by Class")
plt.xlabel("Class")
plt.ylabel("Number of Passengers")

# 3¬†Add number labels
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2, y+5, f"{int(y)}", ha='center')

plt.tight_layout()
plt.show()
```

### Lab‚ÄØ4¬†Pie vs.¬†Bar¬†‚Äì¬†Which Is Clearer?

**Objectives**

- Compare pie and bar charts for the same proportion data
- Discuss readability

```python
import matplotlib.pyplot as plt

fruits = ["Apple", "Banana", "Cherry", "Durian"]
sales  = [30, 45, 15, 10]

fig, ax = plt.subplots(1, 2, figsize=(10,4))

# Pie
ax[0].pie(sales, labels=fruits, autopct="%1.0f%%", startangle=90)
ax[0].set_title("Fruit Sales ‚Äì Pie")

# Bar
ax[1].bar(fruits, sales, color="skyblue")
ax[1].set_title("Fruit Sales ‚Äì Bar")
ax[1].set_ylabel("Units sold")

plt.tight_layout()
plt.show()
# ‚ûú Discussion: Which chart makes differences easiest to judge?
```

### Lab‚ÄØ8¬†Style & Theme¬†‚Äì¬†Changing the Look

**Objectives**

- Apply different `Matplotlib` styles
- Observe how style affects readability

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

for style in ["default", "ggplot", "bmh"]:
    plt.style.use(style)
    plt.figure(figsize=(6,3))
    plt.pllot(x, y, label=f"Style: {style}")
    plt.title("Sine Wave")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

### Lab‚ÄØ9¬†Subplots¬†‚Äì¬†Four‚ÄëPanel Iris Review

**Objectives**

- Create a 2‚ÄØ√ó‚ÄØ2 grid of plots with `plt.subplots`
- Share axes where appropriate

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

fig, axes = plt.subplots(2, 2, figsize=(8,6))
axes = axes.flatten()

for i, feat in enumerate(features):
    axes[i].hist(iris[feat], bins=15, color="teal")
    axes[i].set_title(feat.replace("_", " ").title())

plt.tight_layout()
plt.show()
```
